# DRAW A SCATTER PLOT

```python
# a) Scatter plots are suitable for representing discrete data. In fact, the
# previous test scores are more suitable to be represented by scatter
# plots. You can simply replace plot with scatter to realize the display
# method of scatter plots. Note: scatter needs to explicitly specify X
# and Y information.
fig, ax = plt.subplots(nrows=3, ncols=1, sharex=True)
ax[0].scatter(exam_data.index, exam_data['math score'], color='black', label='math score')
ax[1].scatter(exam_data.index, exam_data['reading score'], color='red', label='reading score')
ax[2].scatter(exam_data.index, exam_data['writing score'], color='green', label='wring score')
ax[0].legend(loc="lower right")
ax[1].legend(loc="upper left")
ax[2].legend(loc="lower center")
```

```python
# b) Scatter plots are also suitable for understanding the correlation
# between two variables by viewing graphics. For example, if we
# want to know that there is a direct correlation between math score
# and reading score, we can try to use the two variables as the X
# and Y axes respectively. to draw and view the drawing results. s
# can control the size of the point, and c can control the color.
fig, ax = plt.subplots()
ax.scatter(exam_data[['math score']], exam_data[['reading score']], s=0.5, c='r', label='math and reading score')
ax.legend(loc='best')
```

```python
# c) It can be clearly seen that there is a positive correlation between math score and reading score. 
# You can try to draw correlation graphs between math score and writing score and reading score and writing score.
fig, ax = plt.subplots()
ax.scatter(exam_data[['math score']], exam_data[['writing score']], s=0.5, c='g', label='math and writing score')
ax.legend(loc='best')

fig, ax = plt.subplots()
ax.scatter(exam_data[['writing score']], exam_data[['reading score']], s=0.5, c='y', label='writing and reading score')
ax.legend(loc='best')
```

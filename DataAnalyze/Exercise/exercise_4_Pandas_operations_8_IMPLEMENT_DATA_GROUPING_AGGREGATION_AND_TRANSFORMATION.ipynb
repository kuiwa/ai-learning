# IMPLEMENT DATA GROUPING , AGGREGATION AND TRANSFORMATION

```python
# a) This part of the exercise is based on d_3 data . 
# First of all, before processing grouped data, you need to know which columns of data are suitable for grouping. 
# You can use info function obtains the attributes of the data and makes the initial judgment.
d_3 = pd.read_csv('exams.csv')
d_3.info()
```

```python
# b) Suppose, first select gender as grouped data, you can pass np.unique function to quickly obtain the unique data information of this column
d_3['gender'].unique()
```

```python
# c) And if you check unique data for data that is not suitable for grouping, you will get a lot of results, so grouping is meaningless.
d_3['math score'].unique()
```

```python
# d) Next, you may need to know the basic information after the data is grouped, such as the grouping situation, that is, which data is
# divided into which group. You can view it based on the groups attribute of the groupby result.
d_3.groupby(['gender']).groups
```

```python
# e) The results show two sets of data based on gender and the value of the data is the Index value. 
# At the same time, you can obtain the grouped result data of the specified group based on get_group, such as the data of the male group:
d_3.groupby(['gender']).get_group('male')
```

```python
# f) The grouped data can be subjected to corresponding aggregation operations. The simplest method is to use the count method to count the quantity.
d_3.groupby(['gender']).count()
```

```python
# g) It can be found that the aggregated result will use gender as an index. 
# If you do not want this to happen, you can add as_index=False during groupby to achieve this.
d_3.groupby(['gender'], as_index=False).count()
```

```python
# h) At the same time, you can see that the value of each column in the output is actually the same. 
# If you only want to see a certain column, such as lunch, you can use [] to specify the specific column name information you want to see:
d_3.groupby(['gender'], as_index=False)['lunch'].count()
```

```python
# i) General aggregation operations are applicable to numeric data, such as sums, averages, and other operations. 
# For example, you want to get the average information of three test scores based on gender grouping:
d_3.groupby(['gender'], as_index=False)[['math score', 'reading score', 'writing score']].count()
```

```python
# j) Of course, applying the describe function to grouped data can obtain information on multiple statistical analysis values at once:
d_3.groupby(['gender'], as_index=False).describe()
```

```python
# k) Use agg Functions can implement more general aggregation tasks. 
# For example, multiple aggregation functions that need to be used can be provided at the same time, 
# such as np.sum , np.mean , to complete multiple aggregation operations at one time.
d_3.groupby(['gender'])[['math score', 'reading score', 'writing score']].agg([np.sum, np.mean])
```

```python
# l) At the same time, you can use the agg function to provide data in dict form and specify different columns to apply different aggregation operations.
d_3.groupby(['gender'])[['math score', 'reading score']].agg({'math score': np.mean, 'reading score': np.std})
```

```python
# m) Different from the aggregation operation, the transform operation performs a separate operation on each grouped data. 
# A simple understanding is that the shape of the grouped data will not change. 
# Use apply you can implement conversion operations with a simple lambda function . 
# For example, if you want to know the difference between the scores of three exams and the average ( lambda x: x- x.mean () ):
d_3[['math score', 'reading score', 'writing score']].apply(lambda x: x - x.mean())
```

```python
# n) When executing apply based on groupby, some warning messages will pop up, indicating that the group_keys parameter needs to be set for groupby. 
# The default value is False, then the grouping key will not appear in the output. 
# Try setting it to True to display the grouping key:
d_3.groupby(['gender'], as_index=True, group_keys=True)[['math score', 'reading score', 'writing score']].apply(lambda x: x - x.mean())
```

# OTHER WAYS TO ACCESS DATA

```python
# a) This part is based on d_4 object, using loc and judgment conditions to implement data access. First, get all rows of gender is male:
d_4.loc[d_4['gender'] == 'male']
```

```python
# b) Get all rows of gender for female and also in group E:
d_4.loc[(d_4['gender'] == 'female') & (d_4['race/ethnicity'] == 'group E')]
```

```python
# c) Get reading score greater than 80 or writing score greater than 80:
d_4.loc[(d_4['reading score'] > 80) | (d_4['writing score'] > 80)]
```

```python
# d) Get all rows of data not in Group E:
d_4.loc[d_4['race/ethnicity'] != 'group E']
```

```python
# e) Get all rows of data not in Group C And gender is female:
d_4.loc[(d_4['race/ethnicity'] != 'group C') & (d_4['gender'] == 'female')]
```

```python
# f) Get all rows with math score in: 40 , 60 , 80 :
d_4.loc[d_4['math score'].isin([40, 60, 80])]
```

```python
# g) Use query function to redo a to f tasks.
# a) This part is based on d_4 object, using loc and judgment conditions to implement data access. First, get all rows of gender is male:
# b) Get all rows of gender for female and also in group E:
d_4.query('gender in "male"'), d_4.query('(gender in "female") and (`race/ethnicity` in "group E")')
```

```python
# b) Get all rows of gender for female and also in group E:
d_4.query('(`reading score` > 80) and (`writing score` > 80)')
```

```python
# d) Get all rows of data not in Group E:
d_4.query('~(`race/ethnicity` in "group E")')
```

```python
# e) Get all rows of data not in Group C And gender is female:
d_4.query('~(`race/ethnicity` in "group C") and (gender in "female")')
```

```python
# f) Get all rows with math score in: 40 , 60 , 80 :
d_4.query('(`math score` in [40, 60, 80])')
```

```python
# h) Based on f The output of the step task is iterated using the itertuples function, and then the value of math score is stored in math_score_list accordingly:
d_6 = d_4.query('(`math score` in [40, 60, 80])')
math_score_list = []
for row in d_6.itertuples():
    math_score_list.append(row[6])
math_score_list
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
exam_data = pd.read_csv("exams.csv")
exam_data.info()
```

```python
fig, ax = plt.subplots()
ax.plot(exam_data["math score"])
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(exam_data["math score"],color="red",linestyle="dotted")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(exam_data["math score"],color="red",linestyle="dotted")
ax.set_xlabel("Index Number")
ax.set_ylabel("Math Score")
ax.set_title("Math Score Plot")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(exam_data[["math score","reading score","writing score"]])
ax.set_xlabel("Index Number")
ax.set_ylabel("Scores")
ax.set_title("Scores Plot")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(exam_data[["math score"]],alpha=0.4)
ax.plot(exam_data[["writing score"]],alpha=0.4)
ax.plot(exam_data[["reading score"]],alpha=0.4)

ax.set_xlabel("Index Number")
ax.set_ylabel("Scores")
ax.set_title("Scores Plot")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(exam_data[["math score"]],alpha=0.4,label="math score")
ax.plot(exam_data[["writing score"]],alpha=0.4,label="writing score")
ax.plot(exam_data[["reading score"]],alpha=0.4,label="reading score")
ax.legend()
ax.set_xlabel("Index Number")
ax.set_ylabel("Scores")
ax.set_title("Scores Plot")
plt.show()
```

```python
fig, ax = plt.subplots(nrows=3,ncols=1)
ax[0].plot(exam_data[["math score"]],color="k")
ax[1].plot(exam_data[["writing score"]],color="r")
ax[2].plot(exam_data[["reading score"]],color="g")
plt.show()
```

```python
fig, ax = plt.subplots(nrows=3,ncols=1)
ax[0].plot(exam_data[["math score"]],color="k",label="math score")
ax[1].plot(exam_data[["writing score"]],color="r",label="writing score")
ax[2].plot(exam_data[["reading score"]],color="g",label="reading score")
ax[0].set_title("math score")
ax[1].set_title("writing score")
ax[2].set_title("reading score")
plt.show()
```

```python
fig, ax = plt.subplots(nrows=3,ncols=1,figsize=(6,10))
ax[0].plot(exam_data[["math score"]],color="k",label="math score")
ax[1].plot(exam_data[["writing score"]],color="r",label="writing score")
ax[2].plot(exam_data[["reading score"]],color="g",label="reading score")
ax[0].set_title("math score")
ax[1].set_title("writing score")
ax[2].set_title("reading score")
plt.show()
```

```python
fig, ax = plt.subplots(nrows=3,ncols=1,sharex=True)
ax[0].plot(exam_data[["math score"]],color="k",label="math score")
ax[1].plot(exam_data[["writing score"]],color="r",label="writing score")
ax[2].plot(exam_data[["reading score"]],color="g",label="reading score")
ax[0].set_title("math score")
ax[1].set_title("writing score")
ax[2].set_title("reading score")
plt.show()
```

```python
fig, ax = plt.subplots(nrows=3,ncols=1,sharex=True,layout='constrained')
ax[0].plot(exam_data[["math score"]],color="k",label="math score")
ax[1].plot(exam_data[["writing score"]],color="r",label="writing score")
ax[2].plot(exam_data[["reading score"]],color="g",label="reading score")
fig.legend(loc="lower right")
plt.show()
```

```python
fig, ax = plt.subplots(nrows=3,ncols=1,sharex=True,layout='constrained')
ax[0].plot(exam_data[["math score"]],color="k",label="math score")
ax[1].plot(exam_data[["writing score"]],color="r",label="writing score")
ax[2].plot(exam_data[["reading score"]],color="g",label="reading score")
ax[0].legend(loc="lower right")
ax[1].legend(loc="upper left")
ax[2].legend(loc="best")
plt.show()
```

# 绘制散点图

```python
fig, ax = plt.subplots(nrows=3,ncols=1,sharex=True,layout='constrained')
ax[0].scatter(exam_data.index,exam_data[["math score"]],color="k",label="math score")
ax[1].scatter(exam_data.index,exam_data[["writing score"]],color="r",label="writing score")
ax[2].scatter(exam_data.index,exam_data[["reading score"]],color="g",label="reading score")
ax[0].legend(loc="lower right")
ax[1].legend(loc="upper left")
ax[2].legend(loc="best")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.scatter(exam_data[["math score"]],exam_data[["reading score"]],s=5,label="math and reading score")
ax.legend(loc="best")
plt.show()
```

# 绘制柱状图

```python
gender_math = exam_data.groupby("gender")["math score"].mean()
fig, ax = plt.subplots()
ax.bar(gender_math.index,gender_math.values)
plt.show()
```

```python
gender_math = exam_data.groupby("gender")["math score"].mean()
fig, ax = plt.subplots()
bar = ax.bar(gender_math.index,gender_math.values)
ax.bar_label(bar,label_type='center')
plt.show()
```

```python
gender = exam_data.groupby("gender")[["math score","reading score","writing score"]].mean()
fig, ax = plt.subplots()
bar = ax.bar(gender.index,gender["math score"].values)
bar = ax.bar(gender.index,gender["reading score"].values)
bar = ax.bar(gender.index,gender["writing score"].values)
ax.bar_label(bar,label_type='center')
ax.set_title("Wrong Bar Sample")
plt.show()
```

```python
gender = exam_data.groupby("gender")[["math score","reading score","writing score"]].mean()
width = 0.25
fig, ax = plt.subplots()
start = 0
for x in range(len(gender.index)):
    bar1 = ax.bar(start+x+width,gender["math score"].values,width)
    bar2 = ax.bar(start+x+width*2,gender["reading score"].values,width)
    bar3 = ax.bar(start+x+width*3,gender["writing score"].values,width)
#     ax.bar_label(bar1,label_type='center')
#     ax.bar_label(bar2,label_type='center')
#     ax.bar_label(bar3,label_type='center')
    start = start+x+width
plt.show()
```

```python
gender = exam_data.groupby("gender")[["math score","reading score","writing score"]].mean()
x = np.arange(len(gender.index))  # the label locations
width = 0.25  # the width of the bars
multiplier = 0
fig, ax = plt.subplots()
for column,content in gender.items():
    offset = width * multiplier
    bar1 = ax.bar(x+offset,content,width,label=column)
    ax.bar_label(bar1)
    multiplier += 1
ax.set_xticks(x + width, gender.index)
    
plt.show()
```

```python
gender_math = exam_data.groupby("gender")["math score"].mean()
fig, ax = plt.subplots()
bar = ax.barh(gender_math.index,gender_math.values,height=0.8)
ax.bar_label(bar,label_type='center')
plt.show()
```

# 绘制饼图

```python
race_grouped_count = exam_data.groupby("race/ethnicity")["gender"].count()
fig, ax = plt.subplots()
ax.pie(race_grouped_count.values,labels=race_grouped_count.index)
plt.show()
```

```python
race_grouped_count = exam_data.groupby("race/ethnicity")["gender"].count()
fig, ax = plt.subplots()
ax.pie(race_grouped_count.values,labels=race_grouped_count.index,autopct="%.1f%%")
plt.show()
```

```python
race_grouped_count = exam_data.groupby("race/ethnicity")["gender"].count()
explode = (0,0,0,0.1,0)
fig, ax = plt.subplots()
ax.pie(race_grouped_count.values,labels=race_grouped_count.index,autopct="%.1f%%",explode=explode)
plt.show()
```

```python
race_grouped_count = exam_data.groupby("race/ethnicity")["gender"].count()
explode = (0,0,0,0.1,0)
fig, ax = plt.subplots()
ax.pie(race_grouped_count.values,labels=race_grouped_count.index,autopct="%.1f%%",explode=explode,shadow=True)
plt.show()
```
